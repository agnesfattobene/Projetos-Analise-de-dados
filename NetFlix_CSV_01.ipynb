```python
# Traduzindo as colunas(Dicionário - PT-BR)
# pip install deep-translator  (outra forma de traduzir: deep-translator)
df.rename(columns={'show_id': 'ID',
                   'type': 'Tipo',
                   'title': 'Título',
                   'director':'Diretor',
                   'cast':'Elenco',
                   'country': 'País',
                   'date_added': 'Data_add',
                   'release_year':'Ano_de_lançamento',
                   'rating':'Avaliação',
                   'duration': 'Duração',
                   'listed_in': 'Listados_em',
                   'description':'Descrição'}, inplace=True)
```

```python
# Calcular a porcentagem de valores nulos em cada coluna
porcentagem_nulos = (df.isnull().sum() / len(df)) * 100

# Exibir a porcentagem de valores nulos em cada coluna
print(round(porcentagem_nulos,1))
```

```python
# Convertendo a coluna de Data_add para datetime
df['Data_add'] = pd.to_datetime(df['Data_add'])
df['Data_add'] = pd.to_datetime(df['Data_add']).astype(int)
rows_with_desconhecido = df[df['Data_add'] == 'Desconhecido'].index
df = df.drop(rows_with_desconhecido)
df['Data_add'] = pd.to_datetime(df['Data_add'])
```

```python
# Convert the 'data_adicionada' column to datetime, setting dayfirst and yearfirst to True
#df['data_adicionada'] = pd.to_datetime(df['data_adicionada'], dayfirst=True, yearfirst=True)
```

```python
# Separando dia, mes e ano em colunas diferentes
df['ano'] = df.Data_add.dt.year

df['mes'] = df.Data_add.dt.month

df['dia'] = df.Data_add.dt.day
```

```python
# Convertendo valores numéricos de float para int
df['Ano_de_lançamento'] = df['Ano_de_lançamento'].astype(int)
```

```python
# Verificando a conversão
df.dtypes
```

```python
# Verificação
df.head(1)
```

```python
# Como é uma grande quantidade de valores NAN nas colunas: Diretor, Elenco e
# País, não vamos excluir os dados NAN, renomeamos NAN para Desconhecido.
df['País'] = df['País'].fillna('Desconhecido')
df['Diretor'] = df['Diretor'].fillna('Desconhecido')
df['Elenco'] = df['Elenco'].fillna('Desconhecido')
df['Avaliação'] = df['Avaliação'].fillna('Desconhecido')
```

```python
# Importando a ferramenta de verificação de qualidade de dados
import pandera as pa
```

```python
# Definição do esquema de validação
schema = pa.DataFrameSchema({'ID': pa.Column(pa.String),
                             'Tipo': pa.Column(pa.String),
                             'Título': pa.Column(pa.String),
                             'Diretor': pa.Column(pa.String),
                             'Elenco': pa.Column(pa.String),
                             'País': pa.Column(pa.String),
                             'Data_add': pa.Column(pa.DateTime),
                             'Ano_de_lançamento': pa.Column(pa.Int),
                             'Avaliação': pa.Column(pa.String),
                             'Duração': pa.Column(pa.String),
                             'Listados_em': pa.Column(pa.String),
                             'Descrição': pa.Column(pa.String),
                             'ano': pa.Column(pa.Int),
                             'mes': pa.Column(pa.Int),
                             'dia': pa.Column(pa.Int)
                            })
```

```python
# Validar o DataFrame
schema.validate(df)
```

```python
# Importando bibliotecas de visualização
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Quais foram os anos com mais filmes lançados em geral?
df.groupby('Ano_de_lançamento')['mes'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.xlabel('Ano_de_lançamento')
plt.ylabel('qtd')
plt.title('Ano_de_lançamento')
plt.show()
```

```python
# Análise de correlação entre as variáveis
correlation = df.corr(method='pearson', numeric_only=True)
mascara = np.triu(np.ones_like(correlation, dtype=bool))
plt.figure(figsize = ((4, 3)))
plot = sns.heatmap(correlation,
                   mask=mascara,
                   annot = True,
                   fmt=".2f", vmax=1, center=0, vmin=-1,
                   cbar=True, cmap='coolwarm',
                   linewidths=.5,
                   cbar_kws={"shrink": .5, 'label': 'Correlação', 'orientation': 'vertical'})
plt.show()
```

```python
# Quais foram os assuntos com menos visualizações em geral?
df.groupby('Diretor')['ano'].sum().sort_values(ascending=True).head(10).plot(kind='barh', color='purple')
plt.xlabel('ano')
plt.ylabel('Tema')
plt.title('Tema com menos Visualizações')
plt.show()
```

```python
# Quais foram os assuntos com menos visualizações em geral?
df.groupby('Tipo')['ano'].sum().sort_values(ascending=True).head(10).plot(kind='barh', color='b')
plt.xlabel('ano')
plt.ylabel('Tema')
plt.title('Tema com menos Visualizações')
plt.show()
```

```python
# Quais foram os assuntos com menos visualizações em geral?
df.groupby('País')['ano'].sum().sort_values(ascending=True).head(10).plot(kind='barh', color='red')
plt.xlabel('ano')
plt.ylabel('Tema')
plt.title('Tema com menos Visualizações')
plt.show()
```
