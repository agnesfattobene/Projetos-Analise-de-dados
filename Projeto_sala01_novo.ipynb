```python
# Função para cadastrar um novo funcionário
def cadastrar_funcionario():
    funcionario = {}  # Dicionário para armazenar os dados do funcionário

    # Solicita os dados do funcionário
    nome = input("Digite o nome do funcionário: ")
    cargo = input("Digite o cargo do funcionário: ")

    # Faz tratamento de exceção para o salário, garantindo que seja um valor válido
    while True:
        try:
            salario = float(input("Digite o salário do funcionário: "))
            break
        except ValueError:
            print("Valor inválido. Digite um número válido para o salário.")

    escolaridade = input("Digite a escolaridade do funcionário: ")
    cidade = input("Digite a cidade do funcionário: ")
    estado = input("Digite o estado do funcionário: ")

    # Faz tratamento de exceção para a idade, garantindo que seja um valor válido
    while True:
        try:
            idade = int(input("Digite a idade do funcionário: "))
            break
        except ValueError:
            print("Valor inválido. Digite um número inteiro válido para a idade.")

    email = input("Digite o e-mail do funcionário: ")

    # Atualiza o dicionário com os dados do funcionário
    funcionario['nome'] = nome
    funcionario['cargo'] = cargo
    funcionario['salario'] = salario
    funcionario['escolaridade'] = escolaridade
    funcionario['cidade'] = cidade
    funcionario['estado'] = estado
    funcionario['idade'] = idade
    funcionario['email'] = email

    # Adiciona o funcionário à lista de funcionários cadastrados
    funcionarios.append(funcionario)

    print("Funcionário cadastrado com sucesso!")


# Função para listar todos os funcionários
def listar_funcionarios():
    print("Lista de funcionários cadastrados:")

    if funcionarios:  # Verifica se a lista de funcionários não está vazia
        for funcionario in funcionarios:
            print("------------------------")
            print("Nome: ", funcionario['nome'])
            print("Cargo: ", funcionario['cargo'])
            print("Salário: ", funcionario['salario'])
            print("Escolaridade: ", funcionario['escolaridade'])
            print("Cidade: ", funcionario['cidade'])
            print("Estado: ", funcionario['estado'])
            print("Idade: ", funcionario['idade'])
            print("E-mail: ", funcionario['email'])
    else:
        print("Nenhum funcionário cadastrado.")


# Função para buscar um funcionário pelo nome
def buscar_funcionario():
    if funcionarios:  # Verifica se a lista de funcionários não está vazia
        nome = input("Digite o nome do funcionário que deseja buscar: ")

        encontrado = False  # Variável para indicar se o funcionário foi encontrado ou não

        for funcionario in funcionarios:
            if funcionario['nome'].lower() == nome.lower():
                print("------------------------")
                print("Nome: ", funcionario['nome'])
                print("Cargo: ", funcionario['cargo'])
                print("Salário: ", funcionario['salario'])
                print("Escolaridade: ", funcionario['escolaridade'])
                print("Cidade: ", funcionario['cidade'])
                print("Estado: ", funcionario['estado'])
                print("Idade: ", funcionario['idade'])
                print("E-mail: ", funcionario['email'])
                encontrado = True
                break

        if not encontrado:
            print("Funcionário não encontrado.")

    else:
        print("Nenhum funcionário cadastrado.")


# Função para atualizar um funcionário pelo nome
def atualizar_funcionario():
    if funcionarios:  # Verifica se a lista de funcionários não está vazia
        nome = input("Digite o nome do funcionário que deseja atualizar: ")

        encontrado = False  # Variável para indicar se o funcionário foi encontrado ou não

        for funcionario in funcionarios:
            if funcionario['nome'].lower() == nome.lower():
                print("------------------------")
                print("Funcionário encontrado! Digite as novas informações:")

                cargo = input("Digite o novo cargo do funcionário: ")

                while True:
                    try:
                        salario = float(input("Digite o novo salário do funcionário: "))
                        break
                    except ValueError:
                        print("Valor inválido. Digite um número válido para o salário.")

                escolaridade = input("Digite a nova escolaridade do funcionário: ")
                cidade = input("Digite a nova cidade do funcionário: ")
                estado = input("Digite o novo estado do funcionário: ")

                while True:
                    try:
                        idade = int(input("Digite a nova idade do funcionário: "))
                        break
                    except ValueError:
                        print("Valor inválido. Digite um número inteiro válido para a idade.")

                email = input("Digite o novo e-mail do funcionário: ")

                # Atualiza as informações do funcionário
                funcionario['cargo'] = cargo
                funcionario['salario'] = salario
                funcionario['escolaridade'] = escolaridade
                funcionario['cidade'] = cidade
                funcionario['estado'] = estado
                funcionario['idade'] = idade
                funcionario['email'] = email

                print("Funcionário atualizado com sucesso!")
                encontrado = True
                break

        if not encontrado:
            print("Funcionário não encontrado.")

    else:
        print("Nenhum funcionário cadastrado.")


# Função para excluir um funcionário pelo nome
def excluir_funcionario():
    if funcionarios:  # Verifica se a lista de funcionários não está vazia
        nome = input("Digite o nome do funcionário que deseja excluir: ")

        removido = False  # Variável para indicar se o funcionário foi removido ou não

        for funcionario in funcionarios:
            if funcionario['nome'].lower() == nome.lower():
                funcionarios.remove(funcionario)
                print("Funcionário removido com sucesso!")
                removido = True
                break

        if not removido:
            print("Funcionário não encontrado.")

    else:
        print("Nenhum funcionário cadastrado.")


# Lista para armazenar os funcionários cadastrados
funcionarios = []

# Loop principal do programa
while True:
    print("------------------------")
    print("1 - Cadastrar funcionário")
    print("2 - Listar funcionários")
    print("3 - Buscar funcionário")
    print("4 - Atualizar funcionário")
    print("5 - Excluir funcionário")
    print("0 - Sair")

    opcao = input("Digite a opção desejada: ")

    if opcao == '1':  # Chama a função para cadastrar um novo funcionário
        cadastrar_funcionario()

    elif opcao == '2':  # Chama a função para listar os funcionários
        listar_funcionarios()

    elif opcao == '3':  # Chama a função para buscar um funcionário pelo nome
        buscar_funcionario()

    elif opcao == '4':  # Chama a função para atualizar um funcionário pelo nome
        atualizar_funcionario()

    elif opcao == '5':  # Chama a função para excluir um funcionário pelo nome
        excluir_funcionario()

    elif opcao == '0':  # Encerra o programa
        print("Encerrando o programa...")
        break

    else:
        print("Opção inválida. Digite uma opção válida.")
```
